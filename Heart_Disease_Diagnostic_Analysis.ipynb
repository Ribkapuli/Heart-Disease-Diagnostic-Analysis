```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
```

```python
d=pd.read_csv('/content/Heart_disease_data.csv')
d.head()
```

```python
n=d.groupby('target').size()
n
```

```python
def heart(h):
  if h==0:
    return 'Absence'
  else:
    return 'Presence'

d['Heart_Disease']=d['target'].apply(heart)
d.head(10)
```

```python
h=d.groupby('Heart_Disease')['target'].count()
h
```

```python
colors = ['#ff9999','#66b3ff']

plt.figure(figsize=(10, 7))
plt.pie(h, labels=['Absence', 'Presence'], autopct='%0.0f%%', colors=colors)
plt.title('Heart Disease Population %', fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)
age = d['age'].values
sex = d['sex'].values

sns.histplot(age, ax=ax[0], color='red', kde=True)
ax[0].set_title('Age Distribution')
ax[0].set_xlim([min(age), max(age)])

sns.histplot(sex, ax=ax[1], color='black', kde=True)
ax[1].set_title('Sex Distribution')
ax[1].set_xlim([min(sex), max(sex)])

plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='age',data=d)
plt.title("Population Age",fontsize=17)
plt.xlabel("Age",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
min_age=d['age'].min()
max_age=d['age'].max()
mean_age=d['age'].mean()
print("Minimum Age = ",min_age)
print("Maximum Age = ",max_age)
print("Meanimum Age = ",mean_age)
```

```python
Young_Ages=d[(d['age']>=29) & (d['age']<40)]
Middle_Ages=d[(d['age']>=40) & (d['age']<55)]
Elderly_Ages=d[(d['age']>=55)]
print("Young Ages=",len(Young_Ages))
print("Middle Ages=",len(Middle_Ages))
print("Elderly Ages=",len(Elderly_Ages))
```

```python
def gender(r):
  if r==1:
    return "Male"
  else:
    return "Female"

d['sex1']=d['sex'].apply(gender)
d.head()
```

```python
def age_range(r):
  if r>=29 and r<40:
    return "Young Age"
  elif r>=40 and r<55:
    return "Middle Age"
  else:
    return "Elder Age"

d['Age_Range']=d['age'].apply(age_range)
d.head()
```

```python
plt.figure(figsize=(10,7))
sns.stripplot(x='Age_Range', y='age', hue='sex1', data=d, order=['Young Age', 'Middle Age', 'Elder Age'], palette='BuGn_r', jitter=True, dodge=True)
plt.title('Gender Based Age Category', fontsize=17)
plt.xlabel('Age Category', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,5))
hue_order=['Young Age','Middle Age','Elder Age']
sns.countplot(x='Heart_Disease',hue='Age_Range',data=d,order=['Presence','Absence'],hue_order=hue_order,palette='rocket')
plt.title('Heart Disease Based on Age Category',fontsize=17)
plt.xlabel('Heart Disease',fontsize=15)
plt.ylabel('Counts',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x=d['Heart_Disease'],hue='sex1',data=d,palette='pastel')
plt.title('Heart Disease Based on Gender',fontsize=17)
plt.xlabel('Heart Disease',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.legend(labels=['Male','Female'])
plt.show()
```

```python
sns.countplot(x=d['Heart_Disease'],hue='cp',data=d,order=['Presence','Absence'])
plt.title('Chest Pain Explained',fontsize=17)
plt.xlabel('Heart Disease',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```

```python
sns.countplot(x=d['sex1'],hue='cp',data=d)
plt.title('Chest Pain Based on Gender',fontsize=17)
plt.xlabel('Sex',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```

```python
sns.countplot(x=d['Age_Range'],hue='cp',data=d,order=['Young Age','Middle Age','Elder Age'],palette='BuGn_r')
plt.title('Chest Pain Based on Age Category',fontsize=17)
plt.xlabel('Age Category',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```
